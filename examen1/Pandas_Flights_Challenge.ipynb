```python
# Your code to explore the data

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


df_flights = pd.read_csv('flights.csv')
df_flights.head()
```

# ejercicio 1

```python

# 1.1 Identifica cualquier null o missing data, e imputa los valores de reemplazo adecuados.

# Calcular el número de valores nulos por columna
valores_nulos = df_flights.isnull().sum()

# Ordenar las columnas por número de valores nulos (de mayor a menor)
valores_nulos_ordenados = valores_nulos.sort_values(ascending=False)

# Mostrar las columnas con valores nulos
print("Columnas con valores nulos:")
print(valores_nulos_ordenados[valores_nulos_ordenados > 0])

# Calcular el porcentaje de valores nulos
porcentaje_nulos = (valores_nulos / len(df_flights)) * 100

# Crear un gráfico de barras para visualizar el porcentaje de valores nulos
plt.figure(figsize=(12, 6))
porcentaje_nulos.sort_values(ascending=False).plot(kind='bar')
plt.title('Porcentaje de Valores Nulos por Columna')
plt.xlabel('Columnas')
plt.ylabel('Porcentaje de Valores Nulos')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# ejercicio 1.1

```python
# 1. comienza limpiando los datos

# Verificar cuántos valores nulos hay en DepDel15 antes
nulos_antes = df_flights['DepDel15'].isnull().sum()
print(f"Valores nulos en DepDel15 antes: {nulos_antes}")

# Actualizar DepDel15 basándose en DepDelay
df_flights['DepDel15'] = (df_flights['DepDelay'] > 15).astype(float)


# Verificar la consistencia entre DepDelay y DepDel15
inconsistencias = ((df_flights['DepDelay'] > 15) != (df_flights['DepDel15'] == 1)).sum()
print(f"Número de inconsistencias entre DepDelay y DepDel15: {inconsistencias}")


# Calcular el porcentaje de vuelos con retraso en la salida
porcentaje_retraso = (df_flights['DepDel15'] == 1).mean() * 100
print(f"\nPorcentaje de vuelos con retraso en la salida: {porcentaje_retraso:.2f}%")

```

# ejercicio 2. 1

```python

# * ¿Cuáles son los retrasos promedio (media) en la salida y llegada?

mean_dep_delay = df_flights['DepDelay'].mean()
mean_arr_delay = df_flights['ArrDelay'].mean()
print(f'Retraso promedio en Salida: {mean_dep_delay:.2f} minutos')
print(f'Retraso promedio en Llegada: {mean_arr_delay:.2f} minutos')



# Datos para el gráfico
delays = [mean_dep_delay, mean_arr_delay]
labels = ['Retraso en Salida', 'Retraso en Llegada']

# Crear el gráfico de barras
plt.bar(labels, delays, color=['blue', 'orange'])

# Añadir títulos y etiquetas
plt.title('Retrasos Promedios en Salida y Llegada')
plt.ylabel('Minutos')

# Mostrar el gráfico
plt.show()
```

# ejercicio 2. 2

```python
# * ¿Cómo se comparan las aerolíneas en términos de desempeño de retrasos en la llegada?
carrier_delays = df_flights.groupby('Carrier')['ArrDelay'].mean()

# Crear el gráfico de barras
carrier_delays.plot(kind='bar', color='lightgreen')

# Añadir títulos y etiquetas
plt.title('Retrasos Promedio por Aerolínea')
plt.xlabel('Aerolínea')
plt.ylabel('Retraso Promedio (minutos)')

# Rotar las etiquetas del eje X para mayor legibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

# ejercicio 2. 3

```python
# * ¿Hay una noticable diferencia en los retrasos de llegada para diferentes días de la semana?

df_flights['DayOfWeek'] = df_flights['DayOfWeek'].astype(int)
day_of_week_delays = df_flights.groupby('DayOfWeek')['ArrDelay'].mean()
print('Retrasos promedio por día de la semana:') 

# Días de la semana (suponiendo que 1 es lunes y 7 es domingo)
days = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Crear el gráfico de barras
plt.bar(days, day_of_week_delays, color='pink')

# Añadir títulos y etiquetas
plt.title('Retrasos Promedios en Llegada por Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Retraso Promedio (minutos)')

# Mostrar el gráfico
plt.show()

# graficar por dias de la semana
```

# ejercicio 2.4

```python
# que aeropuerto de salida tiene el mayor retraso promedio en la salida?


# Agrupar por aeropuerto de salida y calcular el retraso promedio en la salida
airport_delays = df_flights.groupby('OriginAirportName')['DepDelay'].mean()

# Ordenar los retrasos promedio de mayor a menor
airport_delays_sorted = airport_delays.sort_values(ascending=False)

# Mostrar el aeropuerto con el mayor retraso promedio en la salida
print('Aeropuerto de salida con mayor retraso promedio en la salida:')
max_airport = airport_delays_sorted.head(1)
print(max_airport)

# Opcional: Graficar los 10 aeropuertos con mayor retraso promedio
top_10_airports = airport_delays_sorted.head(10)

# Crear etiquetas para los aeropuertos
airport_labels = top_10_airports.index

# Crear el gráfico de barras
plt.barh(airport_labels, top_10_airports, color='lightblue')

# Añadir títulos y etiquetas
plt.title('Top 10 Aeropuertos con Mayor Retraso Promedio en Salida')
plt.xlabel('Retraso Promedio en Salida (minutos)')
plt.ylabel('ID del Aeropuerto')

# Invertir el eje Y para que el aeropuerto con más retraso esté en la parte superior
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()
```

# ejercicio 2. 5

```python
# los despegues tardíos tienden a resultar en retrasos mas largos en la llegada que los desplegues a tiempo?

vuelos_a_tiempo = df_flights[df_flights['DepDelay'] <= 0]
tabla_a_tiempo = pd.crosstab(
    vuelos_a_tiempo['DepDelay'] <= 0,
    vuelos_a_tiempo['ArrDelay'] <= 0,
    normalize='index'
) * 100 # para mostrar en porcentaje

# Tabla 2: Vuelos que despegaron tarde
vuelos_tarde = df_flights[df_flights['DepDelay'] > 0]
tabla_tarde = pd.crosstab(
    vuelos_tarde['DepDelay'] > 0,
    vuelos_tarde['ArrDelay'] <= 0,
    normalize='index'
) * 100 # para mostrar en porcentaje

# Mostrar las tablas
print("Tabla 1: Vuelos que despegaron a tiempo o antes")
print(tabla_a_tiempo.round(2))
print("\nTabla 2: Vuelos que despegaron tarde")
print(tabla_tarde.round(2))


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

tabla_a_tiempo.plot(kind='bar', ax=ax1, width=0.35)
ax1.set_title('Vuelos que despegaron a tiempo o antes')
ax1.set_xlabel('Despegue a tiempo o antes')
ax1.set_ylabel('Porcentaje')
ax1.legend(['Llegada tarde', 'Llegada a tiempo o antes'])

tabla_tarde.plot(kind='bar', ax=ax2, width=0.35)
ax2.set_title('Vuelos que despegaron tarde')
ax2.set_xlabel('Despegue tarde')
ax2.set_ylabel('Porcentaje')
ax2.legend(['Llegada tarde', 'Llegada a tiempo o antes'])

plt.tight_layout()
plt.show()


```

# ejercicio 2. 6

```python
# * ¿Qué ruta (del aeropuerto de origen al aeropuerto de destino) tiene la mayor cantidad de llegadas tardías?

# Filtrar vuelos con llegadas tardías (retraso mayor a 15 minutos)
late_arrivals = df_flights[df_flights['ArrDelay'] > 15]

# Contar el número de llegadas tardías por ruta (aeropuerto de origen y destino)
route_late_arrivals = late_arrivals.groupby(['OriginAirportName', 'DestAirportName']).size()


# Ordenar las rutas por la cantidad de llegadas tardías en orden descendente y seleccionar las 10 principales
top_10_routes = route_late_arrivals.sort_values(ascending=False).head(10)

# Crear etiquetas para las rutas
routes = [f'{origin}->{dest}' for origin, dest in top_10_routes.index]

# Crear el gráfico de barras
plt.barh(routes, top_10_routes, color='skyblue')

# Añadir títulos y etiquetas
plt.title('Top 10 Rutas con Mayor Cantidad de Llegadas Tardías')
plt.xlabel('Cantidad de Llegadas Tardías')
plt.ylabel('Ruta (Origen -> Destino)')

# Invertir el eje Y para que la ruta con más llegadas tardías esté en la parte superior
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()

```

# ejercicio 2. 7

```python

# * ¿Qué ruta tiene el mayor retraso promedio en la llegada?  


# Agrupar por ruta (aeropuerto de origen y destino) y calcular el retraso promedio en la llegada
route_delays = df_flights.groupby(['OriginAirportName', 'DestAirportName'])['ArrDelay'].mean()

# Ordenar las rutas por retraso promedio de mayor a menor
route_delays_sorted = route_delays.sort_values(ascending=False)


# Opcional: Graficar las 10 rutas con mayor retraso promedio
top_5_routes = route_delays_sorted.head(5)

# Crear etiquetas para las rutas con el formato 'Origen -> Destino'
routes = [f'{origin} -> {dest}' for origin, dest in top_5_routes.index]

# Crear el gráfico de barras
plt.barh(routes, top_5_routes, color='orange')

# Añadir títulos y etiquetas
plt.title('Top 5 Rutas con Mayor Retraso Promedio en Llegada')
plt.xlabel('Retraso Promedio (minutos)')
plt.ylabel('Ruta (Origen -> Destino)')

# Invertir el eje Y para que la ruta con más retraso esté en la parte superior
plt.gca().invert_yaxis()

# Mostrar el gráfico
plt.show()

```
